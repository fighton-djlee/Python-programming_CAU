## <mark style="background-color: #d3d3de;"> Typical Pitfalls of For Loops </mark>

### For loop 안에서 iterable을 삭제할 때 생기는 문제
- 이터레이터가 순회할 때, 인덱싱에서 점프가 생기면서 필요한 동작이 이루어지지 않는 경우 

```python
numbers = [2, 4, 6, 8]

for value in numbers:
  if value % 2 == 0:
    numbers.remove(value)

print(numbers)
```

##### 해결책 #1
- 주어진 iterable을 그대로 두고, 같은 값을 갖는 별도의 iterable 객체를 대상으로 이터레이터 프로토콜을 사용. 
- 리스트 전체 크기와 동일한 리스트 객체가 또 하나 생성되므로 메모리 사용 부담

```python
numbers = [2, 4, 6, 8]

for value in numbers[::-1]:
  if value %2 == 0:
    numbers.remove(value)

print(numbers)
```

#### 해결책 #2
- 주어진 iterable을 역순으로 순회하는 이터레이터 프로토콜을 사용. 
- 뒤쪽에서부터 삭제하면서 인덱싱이 정상적으로 유지. 
- 실제로 역순의 리스트 객체가 생성되지는 않음 (메모리 사용 측면에서 유리)

```python
numbers = [2, 4, 6, 8]

for value in reversed(numbers):
  if value %2 == 0:
    numbers.remove(value)

print(numbers)
```

## <mark style="background-color: #d3d3d3;">알고리즘 구현 예제 (제곱근 구하기 from Lecture 1) </mark>

### \[ **<u>알고리즘 리뷰</u>** \]
- 변수 목록
  1. 제곱근을 구하고자 하는 값 입력 with input 함수: x_number
  2. 제곱근 값의 최초 추측값 입력: sqrt_guess (계속 업데이트)
  3. sqrt_guess의 제곱과 x_number의 차이 (절대값): abs_err
  4. sqrt_guess의 제곱이 x_number와 충분히 가까운지 판단할 Boolean 변수: is_close
  5. 기타
      - steps: 루프 반복 회수
      - ans: math sqrt 함수로 구한 정확한 값 (알고리즘으로 구한 값과의 차이 계산 목적)
- 알고리즘
  <ul>
      <li> abs_err 계산하여 x_number와 가까운지 비교 (예: abs_err < 1e-10)하여 is_close에 할당 </li>
      <li> is_close가 False인 경우, sqrt_guess를 업데이트한 후 (현 단계 번호와 abs_err 값 출력) while loop 계속 반복 </li>
      <li> is_close가 True로 판정되는 순간, 최종 abs_err값 출력 후 break하여 루프 종료</li>
      <li> 계산된 제곱근 값 sqrt_guess print. 비교 목적으로 sqrt 함수로 구한 값과의 차이도 print</li>
  </ul>

### While...break 구현 방법

```python
import math
your_number = input(f"Enter a positive number you'd like to get square root of: ")
x_number = float(your_number)
sqrt_guess = float(input(f"Give me an initial guess for the square root: "))
steps = 0

while True:
    is_close_enough = abs(abs_err := abs(sqrt_guess **2 - x_number)) < 1e-12
    if not is_close_enough:
        sqrt_guess = (sqrt_guess + x_number/sqrt_guess)/2.0
        steps += 1
        print(f"{steps} steps have been executing. Current error is {abs_err: .10f}")
    else:
        print(f"{steps+1} steps have been executed. Final error is {abs_err: .10f}")
        break

print(f"\nCalculation completed. The square root of {x_number} is {sqrt_guess: .16f}.")
print(f"\nAnswer from math module is {(ans := math.sqrt(x_number))} \
and the difference is {abs(ans-sqrt_guess):.16f}.")
```

### While True와 if 비교구문 통합 구현

```python
import math
your_number = input(f"Enter a positive number you'd like to get square root of: ")
x_number = float(your_number)
sqrt_guess = float(input(f"Give me an initial guess for the square root: "))

while not( (abs_error := abs(sqrt_guess**2 - x_number)) < 1e-12 ) :
    print(f"Error at step is {abs_error: .16f}")
    sqrt_guess = (sqrt_guess + x_number/sqrt_guess)/2.0
    
print(f"\nCalculation completed. The square root of {x_number} is {sqrt_guess: .16f}.")
print(f"\nAnswer from math module is {(ans := math.sqrt(x_number))} \
and the difference is {abs(ans-sqrt_guess): .16f}.")
```

## <mark style="background-color: #d3d3de;">While 루프 활용 예제: 숫자 맞히기 게임 </mark>

- LOW~HIGH 범위에서 *randint* 함수로 발생한 난수(random number)를 힌트를 통해 추측해서 맞히는 과정을 구현
- 변수 정의
  1. LOW, HIGH: 난수의 최소, 최대값 범위. 
  2. secret_number: 플레이어가 맞혀야 하는 난수의 값
  3. guess/number: 플레이어가 input 함수로 입력하는 추측값 / 정수형 변환값
  4. clue: 추측값과 정답을 비교하여 터미널로 출력하기 위한 힌트 문자열 (크다, 적다)

##### 무한 While loop

```python
from random import randint

LOW, HIGH = 1, 99

secret_number = randint(LOW, HIGH)
clue = ""

while True:
    guess = input(f"Guess a number between {LOW} and {HIGH} {clue} ")
    number = int(guess)
    if number > secret_number:
        clue = f"(less than {number})"
    elif number < secret_number:
        clue = f"(greater than {number})"
    else:
      print(f"You nailed it! The secret number is {number}")
      break
```

##### 시도 회수에 제한을 두는 경우
- 누적된 시도 회수가 최대 시도회수보다 적을 경우에만 루프 수행
- 정답을 찾지 못하고 (break 탈출 없이) 최대 시도 회수를 넘기면 루프가 종료되고, else 블록이 수행됨

```python
from random import randint

LOW, HIGH = 1, 99
MAX_TRIES = 7

secret_number = randint(LOW, HIGH)
clue = ""
attempts = 0

while attempts <= MAX_TRIES:
    guess = input(f"Guess a number between {LOW} and {HIGH} {clue} ")
    number = int(guess)
    attempts += 1

    if number > secret_number:
        clue = f"(less than {number})"
    elif number < secret_number:
        clue = f"(greater than {number})"
    else:
      print(f"You nailed it! The secret number is {number}")
      break
else:
    print(f"All attempts failed. The secret number is {secret_number}")
```

## <mark style="background-color: #d3d3d3;"> Mini 실습과제 2번 문제 풀이 </mark>

### 4학년 엉아 모범답안 - 마지막 자리부터 더해 나가기

```python
four_digit_num = int(input(f"4자리 수를 입력하세요: "))
digit_sum = 0

for _ in range(4):
    digit_sum += four_digit_num % 10
    four_digit_num //= 10

print(f"각 자리 수의 합: {digit_sum}")
```

### 첫 자리부터 계산해서 더해 나가는 방법 1

```python
four_digit_str = input("4자리 정수를 입력하세요: ")
four_digit_num = int(four_digit_str)
digit_sum = 0

for idx in range(4):
    digit_sum += four_digit_num // (10 ** (3-idx))
    four_digit_num %= (10 ** (3-idx))

print(f"입력하신 숫자는 {int(four_digit_str)}입니다. 각 자릿수의 합 = {digit_sum}")
```

### 첫 자리부터 계산해서 더해 나가는 방법 2

```python
four_digit_int = int(input("4자리 정수를 입력하세요: "))
digit_sum = 0

for idx in range(4):
    digit_sum += ( four_digit_int // (10 ** (3-idx)) % 10 )

print(f"입력하신 숫자는 {four_digit_int}입니다. 각 자릿수의 합 = {digit_sum}")
```

### 리스트 메소드 append()와 sum() 함수 이용

```python
four_digit_num = input(f"4자리 수를 입력하세요: ")
tmp_lst = []

for char in four_digit_num:
    tmp_lst.append(int(char))

print(f"입력하신 숫자는 {four_digit_num}입니다. 각 자리 수의 합: { (digit_sum := sum(tmp_lst) )}")
```

### 리스트 컴프리헨션으로 한번에 해결하기 (We will cover this later)
